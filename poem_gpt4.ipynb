```python
w=open("titles.txt",'r',encoding='utf-8')
lines=w.readlines()
w.close()

outfile=open("gpt-4.txt",'w',encoding='utf-8')
for line in lines:
    if len(line.strip())>1:
        lst=line.strip()
        completion = client.chat.completions.create(
          model="gpt-4-1106-preview",
          messages=[
            {"role": "system", "content": "你是一名大诗人。请根据给定的标题创作格律诗。以下是一些示例：\n 五言绝句：\n 静夜思 \n 床上明月光，疑是地上霜。\n 举头望明月，低头思故乡。\n 五言律诗：\n 赋得古原草送别 \n离离原上草，一岁一枯荣。\n 野火烧不尽，春风吹又生。\n 远芳侵古道，晴翠接荒城。\n 又送王孙去，萋萋满别情。\n七言绝句：\n 望庐山瀑布 \n 日照香炉生紫烟，遥看瀑布挂前川。\n 飞流直下三千尺，疑是银河落九天。\n 七言律诗：\n黄鹤楼 \n 故人西辞黄鹤楼，烟花三月下扬州。\n 黄鹤一去不复返，白云千载空悠悠。\n晴川历历汉阳树，芳草萋萋鹦鹉洲。\n 日暮乡关何处是，烟波江上使人愁。"},
            {"role": "user", "content": "请选择一种体裁，以标题“"+lst+"”创作一首格律诗。"}
          ]
        )
        print(lst,file=outfile)
        print(lst)
        print(completion.choices[0].message.content)
        print(completion.choices[0].message.content,file=outfile)
outfile.close()
```
